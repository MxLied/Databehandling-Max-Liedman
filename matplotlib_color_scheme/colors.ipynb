```python
### PLOT
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
data_lables = [ 'cases_2020', 'Vacinations 2020', 'Dead 2020',
                'cases_2021', 'Vacinations 2021', 'Dead 2021', 
                'cases_2022', 'Vacinations 2022', 'Dead 2022']
data    = [ 1,2,3, 
            4,5,6, 
            7,8,9]
ax.set_yscale('log')
color_green = [0.161, 0.505, 0.35]
color_blue = [0.161, 0.35, 0.505] 
color_red = [0.48, 0.2, 0.2] 

color_cases = color_green
color_vacc = color_blue
color_dead = color_red 

barplot = ax.bar(data_lables, data, color=[ color_cases, color_vacc, color_dead,
                                            color_cases, color_vacc, color_dead,
                                            color_cases, color_vacc, color_dead])
plt.bar_label(barplot)
plt.title("Sick, Vacinated And Dead")

plt.show()
```
